```python
import numpy as np
import time
import os 
import matplotlib.pyplot as plt
import pandas as pd
import datetime as datetime
from yahoofinancials import YahooFinancials
```

```python
stocks = pd.read_csv('symbols.csv')
stocks = stocks.iloc[:,0].tolist()
```

```python
def download_csv_data(ticker, start_date, end_date, freq):
    if not os.path.exists('/root/.zipline/yahoofiance/daily/'):
        os.makedirs('/root/.zipline/yahoofiance/daily/')

    yahoo_financials = YahooFinancials(ticker=stock)

    df = yahoo_financials.get_historical_price_data(start_date, end_date, freq)
    df = pd.DataFrame(df[ticker]['prices']).drop(['date'], axis=1) \
            .rename(columns={'formatted_date':'date'}) \
            .loc[:, ['date','open','high','low','close','volume']] \
            .set_index('date')
    df.index = pd.to_datetime(df.index)
    df['dividend'] = 0
    df['split'] = 1

    # save data to csv for later ingestion
    df.to_csv('/root/yahoo_data/daily/{}.csv'.format(stock), header=True, index=True)

    # plot the time series
   # df.close.plot(title='{} prices --- {}:{}'.format(ticker, start_date, end_date));

```

```python
start_date='1990-01-02'
end_date=datetime.date.today().isoformat()
```

```python
for stock in stocks:
    try:
        download_csv_data(ticker=stock,start_date=start_date,end_date=end_date ,freq='daily')
        print(stock)
    except Exception as ex:
        print(ex)
```
